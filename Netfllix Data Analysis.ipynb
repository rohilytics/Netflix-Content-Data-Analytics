```python
import matplotlib.pyplot as plt
crosstab = pd.crosstab(df['type'], df['rating'])
crosstab.plot(kind='bar', stacked=True, width=0.8)
plt.title('Content Type vs. Rating')

plt.legend(title='Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
sns.heatmap(pd.crosstab(df['release_year'], df['rating']), cmap='viridis')
plt.title('Heatmap of Release Year vs. Rating')
plt.show()

```
